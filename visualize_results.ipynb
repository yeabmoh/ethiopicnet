```python
from google.colab import drive
import pandas as pd
drive.mount('/content/drive')
%cd "/content/drive/My Drive/neuro140fp"
path = "../training_results" #relative path if running from git
dense_net_path = "DenseNet_training_results.csv"
alex_net_path = "AlexNet_training_results.csv"
vgg_net_path = "VGG_net_training_results.csv"
res_net_path = "ResNet_training_results.csv"

dense_net_data = pd.read_csv("training_results/"+dense_net_path)
alex_net_data = pd.read_csv("training_results/"+alex_net_path)
vgg_net_data = pd.read_csv("training_results/"+vgg_net_path)
res_net_data = pd.read_csv("training_results/"+res_net_path)

dense_net_data.head(), alex_net_data.head(), vgg_net_data.head(), res_net_data.head()
```

#Plotting train and validation accuracies for each model

DenseNet

```python
import matplotlib.pyplot as plt
def plot_model_accuracy(data, title):
    plt.figure(figsize=(8, 6))
    plt.plot(data['epoch'], data['train_accuracy'], label='Training Accuracy', marker='o')
    plt.plot(data['epoch'], data['val_accuracy'], label='Validation Accuracy', marker='o')
    plt.title(f'Training and Validation Accuracy for {title}')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_model_accuracy(dense_net_data, "DenseNet")
```

Alexnet

```python
plot_model_accuracy(alex_net_data, "AlexNet")
```

```python
plot_model_accuracy(vgg_net_data, "VGGNet")
```

```python
plot_model_accuracy(res_net_data, "ResNet")
```

##Plotting all model accuracies

```python
def plot_all_models_accuracy():
    plt.figure(figsize=(8, 7))

    plt.plot(alex_net_data['epoch'], alex_net_data['train_accuracy'], 'r-', label='AlexNet Train Accuracy', marker='o')
    plt.plot(vgg_net_data['epoch'], vgg_net_data['train_accuracy'], 'b-', label='VGGNet Train Accuracy', marker='o')
    plt.plot(res_net_data['epoch'], res_net_data['train_accuracy'], 'g-', label='ResNet Train Accuracy', marker='o')
    plt.plot(dense_net_data['epoch'], dense_net_data['train_accuracy'], 'y-', label='DenseNet Train Accuracy', marker='o')

    plt.plot(alex_net_data['epoch'], alex_net_data['val_accuracy'], 'r--', label='AlexNet Val Accuracy', marker='o')
    plt.plot(vgg_net_data['epoch'], vgg_net_data['val_accuracy'], 'b--', label='VGGNet Val Accuracy', marker='o')
    plt.plot(res_net_data['epoch'], res_net_data['val_accuracy'], 'g--', label='ResNet Val Accuracy', marker='o')
    plt.plot(dense_net_data['epoch'], dense_net_data['val_accuracy'], 'y--', label='DenseNet Val Accuracy', marker='o')

    plt.title('Training and Validation Accuracy Comparison Among Models')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_all_models_accuracy()
```

##Plotting loss for all models

```python
def plot_all_models_loss():
    plt.figure(figsize=(8, 7))

    plt.plot(alex_net_data['epoch'], alex_net_data['train_loss'], 'r-', label='AlexNet Train Loss', marker='o')
    plt.plot(vgg_net_data['epoch'], vgg_net_data['train_loss'], 'b-', label='VGGNet Train Loss', marker='o')
    plt.plot(res_net_data['epoch'], res_net_data['train_loss'], 'g-', label='ResNet Train Loss', marker='o')
    plt.plot(dense_net_data['epoch'], dense_net_data['train_loss'], 'y-', label='DenseNet Train Loss', marker='o')

    plt.plot(alex_net_data['epoch'], alex_net_data['val_loss'], 'r--', label='AlexNet Val Loss', marker='o')
    plt.plot(vgg_net_data['epoch'], vgg_net_data['val_loss'], 'b--', label='VGGNet Val Loss', marker='o')
    plt.plot(res_net_data['epoch'], res_net_data['val_loss'], 'g--', label='ResNet Val Loss', marker='o')
    plt.plot(dense_net_data['epoch'], dense_net_data['val_loss'], 'y--', label='DenseNet Val Loss', marker='o')

    plt.title('Training and Validation Loss Comparison Among Models')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_all_models_loss()
```
